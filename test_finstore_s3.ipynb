```python
from finstore.finstore import Finstore
import pandas as pd
import numpy as np
```

```python
f = Finstore(market_name="crypto_binance", timeframe="4h")
```

# Creating Mock Data

```python
timestamps = pd.date_range('2025-01-01', periods=24, freq='h')
df_sample = pd.DataFrame({
    'timestamp': timestamps,
    'open': np.random.rand(24),
    'high': np.random.rand(24),
    'low': np.random.rand(24),
    'close': np.random.rand(24),
    'volume': np.random.randint(100, 1000, size=24)
})
```

```python
df_sample
```

# Storing, Fetching, Deleting Data

```python
f.S3_mode.store_symbol_s3(symbol='BTC/USDT', df=df_sample)
```

```python
f.S3_mode.fetch_symbol_s3(symbol='BTC/USDT')
```

```python
f.S3_mode.delete_symbol_s3(symbol='BTC/USDT')
```
